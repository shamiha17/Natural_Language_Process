# Disaster Tweet Classification

This dataset contains tweets labeled as real disaster (1) or not (0).

## Columns:
- id: unique identifier
- keyword: disaster-related keyword
- location: tweet location
- text: tweet content
- target: classification label

```python
df_train=pd.read_csv("/kaggle/input/competitions/nlp-getting-started/train.csv")
```

```python
df_test=pd.read_csv("/kaggle/input/competitions/nlp-getting-started/test.csv")
```

```python
df_train
```

```python
df_train.isnull().any()
```

```python
df_train['keyword']=df_train['keyword'].fillna('unknown')

df_train['location'] = df_train['location'].fillna('unknown')
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z]", " ", text)
    return text

df_train['clean_text'] = df_train['text'].apply(clean_text)
df_test['clean_text'] = df_test['text'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df_train['clean_text'])
y = df_train['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

```python
X_test_real = vectorizer.transform(df_test['clean_text'])
```

```python
y_pred = model.predict(X_test_real)
```

```python
import pandas as pd

results = pd.DataFrame({
    'id': df_test['id'],
    'target': y_pred
})
```

```python
results.to_csv('predictionss.csv', index=False)
```
